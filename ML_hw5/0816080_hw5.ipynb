# Data Preprocessing

```python
df_shuffled=sklearn.utils.shuffle(df)
```

```python
def generate_text(data):
    text_data = [" ".join(doc['ingredients']).lower() for doc in data]
    return text_data 
def read_dataset(path):
    return json.load(open(path)) 
train = read_dataset('C:/Users/Suyin/Desktop/ML_hw5/input/train.json')
test = read_dataset('C:/Users/Suyin/Desktop/ML_hw5/input/test.json')
train_text = generate_text(train)
test_text = generate_text(test)
target = [doc['cuisine'] for doc in train]
tfidf = TfidfVectorizer(binary=True)
def tfidf_features(txt, flag):
    if flag == "train":
        x = tfidf.fit_transform(txt).toarray()
    else:
        x = tfidf.transform(txt).toarray()
    x = x.astype('float16')
    return x 
x = tfidf_features(train_text, flag="train")
sub_x = tfidf_features(test_text, flag="test")
y=np.array(target)
lb = LabelEncoder()
y = lb.fit_transform(target)
#y = pd.get_dummies(target).values
```

```python
#Holdout validation with the ratio 7:3
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
y_train = tf.keras.utils.to_categorical(y_train, 20)
y_test = tf.keras.utils.to_categorical(y_test, 20)
```

```python
table=prettytable.PrettyTable()
table.field_names=['ANN ','Accuracy','Recall','Precision']
```

# Artificial Neural Networks 

```python
numfeat = x_train.shape[1]

model = keras.models.Sequential()
model.add(Dense(300,input_dim = numfeat,activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(500,activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(400,activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(20,activation='softmax'))
model.compile(loss = 'categorical_crossentropy',optimizer = 'rmsprop',metrics = ['categorical_accuracy'])
model.fit(x_train, y_train, validation_data=(x_test,y_test),epochs=20,shuffle=True, verbose =2,batch_size=128)
```

```python
print("Accuracy with KERAS" ,model.evaluate(x_test,y_test)[1])
```

# Results 

```python
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test.argmax(axis=1), y_pred.argmax(axis=1))

plt.figure(figsize=(10,10))
f, ax = plt.subplots(figsize =(5,5))
sns.heatmap(cm,annot = True,linewidths=0.2,linecolor="red",fmt = ".0f",ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.title("Confusion Matrix")
plt.show()
```

```python
acc=round(accuracy_score(y_test.argmax(axis=1), y_pred.argmax(axis=1))*100,4)
recall=round(recall_score(y_test.argmax(axis=1), y_pred.argmax(axis=1),average='weighted')*100,4)
prec=round(precision_score(y_test.argmax(axis=1), y_pred.argmax(axis=1),average='weighted')*100,4)

table.add_row(["Keras Sequential Model",acc,recall,prec])
```
